```python
import pandas as pd
df = pd.read_excel("Boston Housing Dataset 1978.xlsx")
df.head()
```

Limpieza, quitar duplicados

```python
data=df.dropna().drop_duplicates()
```

```python
import statsmodels.api as sm
X = data[["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT"]]
X = sm.add_constant(X)
Y = data["MEDV"]
ols = sm.OLS(Y, X)
results = ols.fit()
print(results.summary())
```

```python
data.describe()
```

Las betas no son comparables porque no estan estandarizados.

```python
for col in data.columns.drop("MEDV"):  # todas las columnas excepto MEDV
    col_min = data[col].min()
    col_max = data[col].max()
    data[col] = (data[col] - col_min) / (col_max - col_min)

display(data.describe())
```

```python
import statsmodels.api as sm
X = data[["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT"]]
X = sm.add_constant(X)
Y = data["MEDV"]
ols = sm.OLS(Y, X)
results = ols.fit()
print(results.summary())
```

El modelo de regresin lineal mltiple, nos indica un r^2 alto lo que indica que captura una gran proporcin de la varianza. Observamos que la mayora de las variables son significativas.
