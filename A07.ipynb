```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
X = df.drop('sales', axis=1)
y = df['sales']
```

#Paso 1

```python
model = LinearRegression()
model.fit(X, y)
r2 = model.score(X,y)
print(r2)
```

#Paso 2

```python
modelT = DecisionTreeRegressor()
modelT.fit(X, y)
preds = modelT.predict(X)
r2 = modelT.score(X,y)
print(r2)
```

#Paso 3

En el árbol de decisión el modelo arrojó un valor de 1, lo que indica que el modelo prácticamente se ajusta de manera perfecta a los puntos, por otro lado la regresión lineal nos dio 0.89 lo que significa un ajuste bueno. Puede que el árbol de decisión sobreajusto, mientras que la regresión si generaliza de manera más suave.

Dice el profe que siempre antes de hacer train y test con el árbol de decisión siempre dara 1, aahh tiene sentido ya que se ajuta a todos los datos perfecto siempre.

#Paso 4

```python
X_train,X_test, Y_train,Y_test = train_test_split(X, y, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
r2 = model.score(X_test, Y_test)
print('r2 TEST:', r2)
r2 = model.score(X_train, Y_train)
print('r2 TRAIN:', r2)
```

```python
modelT = DecisionTreeRegressor()
modelT.fit(X_train, Y_train)
preds = modelT.predict(X_test)
r2 = modelT.score(X_test,Y_test)
print('R2 TEST:', r2)
r2 = modelT.score(X_train,Y_train)
print('R2 TRAIN:', r2)
```

#Paso 5

```python
model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model.fit(X, y)
r2 = model.score(X,y)
print(r2)
```

```python
model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model.fit(X_train, Y_train)

r2_test = model.score(X_test, Y_test)
print('r2 TEST:', r2_test)

r2_train = model.score(X_train, Y_train)
print('r2 TRAIN:', r2_train)
```

```python
modelT = make_pipeline(PolynomialFeatures(degree=2), DecisionTreeRegressor())
modelT.fit(X, y)

preds = modelT.predict(X)

r2 = modelT.score(X, y)
print('r2:', r2)
```

```python
modelT = make_pipeline(PolynomialFeatures(degree=2), DecisionTreeRegressor())
modelT.fit(X_train, Y_train)
preds = modelT.predict(X_test)
r2 = modelT.score(X_test,Y_test)
print('R2 TEST:', r2)
r2 = modelT.score(X_train,Y_train)
print('R2 TRAIN:', r2)
```

#Conclusión
En conclusión, los árboles de decisión tienen la capacidad de ajustarse muy fácilmente a todos los datos del conjunto de entrenamiento, lo que puede llevar a valores de R² cercanos a 1.00.
Por esa razón es importante separar el dataset en train y test, para evaluar si el modelo realmente generaliza o solo aprende los datos vistos. En cambio, la regresión lineal normalmente generaliza de forma más estable, y aunque su R² sea más bajo en entrenamiento, suele tener un desempeño más equilibrado entre train y test.
