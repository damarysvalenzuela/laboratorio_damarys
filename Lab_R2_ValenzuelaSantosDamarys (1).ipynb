## Regresión polinomial

Observa la siguiente gráfica.

```python
import matplotlib.pyplot as plt; import numpy as np
%matplotlib inline
np.random.seed(0)
x = np.linspace(-3, 3, 100); y = 1 + 4*x**2 + np.random.normal(0, 1, 100)
plt.figure(); plt.xlabel("x"); plt.ylabel("y")
plt.scatter(x, y); plt.grid()
```

Utiliza `sklearn` para realizar una regresión lineal simple ($\hat{y} = \hat{\beta_0} + \hat{\beta_1}x $)

```python
X = x.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)
```

```python
b0 = modelo.intercept_
b1 = modelo.coef_[0]
print(b0,b1)
```

Crea una gráfica donde se muestren las observaciones como una nube de puntos y la línea que ajustaste con regresión. Calcula el métrico $R^2$.

```python
from sklearn.metrics import r2_score
r2 = r2_score(y, y_pred)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color="red")
plt.grid()
plt.show()
```

Podemos ver que el ajuste no es correcto, pues parece que tiene un comportamiento cuadrático y estamos intentando ajustar una línea recta.

En lugar de hacer una regresión lineal simple, quizás podamos extender el modelo para que tenga varios factores dentro:

$$ Y = \beta_0 + \beta_1 X + \beta_2 X^2 + ... + \beta_p X^p + \epsilon$$

Ajusta un modelo cuadrático con `sklearn` ($\hat{y} = \hat{\beta_0} + \hat{\beta_1}x + \hat{\beta_1}x^2 $) y calcula su $R^2$.

```python
X_quad = np.column_stack((X, X**2))
```

```python
modelo = LinearRegression()
modelo.fit(X_quad, y)
```

```python
y_pred = modelo.predict(X_quad)
r2 = r2_score(y, y_pred)
```

```python
print(modelo.coef_ , modelo.intercept_ , r2)
```

Crea una gráfica con las observaciones y la nueva curva que ajustaste.

```python
plt.scatter(x,y)
plt.plot(x,y_pred,color="red")
plt.grid()
plt.show()
```

Ajusta modelos de grados 3, 5, 7, 9, y 11. Calcula su $R^2$.

Crea una gráfica por cada modelo con las observaciones y su curva.

```python
X3 = np.column_stack((X, X**2,X**3))
modelo = LinearRegression()
modelo.fit(X3, y)
y_pred = modelo.predict(X3)
r2 = r2_score(y, y_pred)
print(r2)
plt.scatter(x,y)
plt.plot(x,y_pred,color="red")
plt.title("Regresión X^3")
plt.grid()
plt.show()
```

```python
X5 = np.column_stack((X, X**2, X**3, X**4, X**5))
modelo = LinearRegression()
modelo.fit(X5, y)
y_pred = modelo.predict(X5)
r2 = r2_score(y, y_pred)
print(r2)
plt.scatter(x, y)
plt.plot(x, y_pred, color="red")
plt.title("Regresión X^5")
plt.grid()
plt.show()
```

```python
X7 = np.column_stack((X, X**2, X**3, X**4, X**5, X**6, X**7))
modelo = LinearRegression()
modelo.fit(X7, y)
y_pred = modelo.predict(X7)
r2 = r2_score(y, y_pred)
print(r2)
plt.scatter(x, y)
plt.plot(x, y_pred, color="red")
plt.title("Regresión X^7")
plt.grid()
plt.show()
```

```python
X9 = np.column_stack((X, X**2, X**3, X**4, X**5, X**6, X**7, X**8, X**9))
modelo = LinearRegression()
modelo.fit(X9, y)
y_pred = modelo.predict(X9)
r2 = r2_score(y, y_pred)
print(r2)
plt.scatter(x, y)
plt.plot(x, y_pred, color="red")
plt.title("Regresión X^9")
plt.grid()
plt.show()
```

```python
X11 = np.column_stack((X, X**2, X**3, X**4, X**5, X**6, X**7, X**8, X**9, X**10, X**11))
modelo = LinearRegression()
modelo.fit(X11, y)
y_pred = modelo.predict(X11)
r2 = r2_score(y, y_pred)
print(r2)
plt.scatter(x, y)
plt.plot(x, y_pred, color="red")
plt.title("Regresión X^11")
plt.grid()
plt.show()
```

**¿Cuál modelo consideras que es mejor? ¿Por qué?**

El modelo cuadrático (grado 2) es el mejor porque se ajusta bien a los datos y captura la relación real entre x,y. Los modelos de mayor grado no mejoran mucho el R^2
 y solo terminan sobreajustando, aprendiendo el ruido en lugar de la tendencia.

Recuerda que una de las metas de ajustar modelos es para poder interpolar y extrapolar. Los datos que tenemos mantienen $x \in [-3, 3]$. ¿Qué pasaría si quisiéramos saber el valor de $y$ cuando $x=5$?

Calcula la predicción $\hat{y}$ para todos los modelos creados (grados 2, 3, 5, 7, 9, y 11).

```python
x_extrap = np.array([[5]])

X2 = np.column_stack((X, X**2))
modelo2 = LinearRegression().fit(X2, y)
y2 = modelo2.predict(np.column_stack((x_extrap, x_extrap**2)))

X3 = np.column_stack((X, X**2, X**3))
modelo3 = LinearRegression().fit(X3, y)
y3 = modelo3.predict(np.column_stack((x_extrap, x_extrap**2, x_extrap**3)))

X5 = np.column_stack((X, X**2, X**3, X**4, X**5))
modelo5 = LinearRegression().fit(X5, y)
y5 = modelo5.predict(np.column_stack((x_extrap, x_extrap**2, x_extrap**3, x_extrap**4, x_extrap**5)))

X7 = np.column_stack((X, X**2, X**3, X**4, X**5, X**6, X**7))
modelo7 = LinearRegression().fit(X7, y)
y7 = modelo7.predict(np.column_stack((x_extrap, x_extrap**2, x_extrap**3, x_extrap**4, x_extrap**5, x_extrap**6, x_extrap**7)))

X9 = np.column_stack((X, X**2, X**3, X**4, X**5, X**6, X**7, X**8, X**9))
modelo9 = LinearRegression().fit(X9, y)
y9 = modelo9.predict(np.column_stack((x_extrap, x_extrap**2, x_extrap**3, x_extrap**4, x_extrap**5, x_extrap**6, x_extrap**7, x_extrap**8, x_extrap**9)))

X11 = np.column_stack((X, X**2, X**3, X**4, X**5, X**6, X**7, X**8, X**9, X**10, X**11))
modelo11 = LinearRegression().fit(X11, y)
y11 = modelo11.predict(np.column_stack((x_extrap, x_extrap**2, x_extrap**3, x_extrap**4, x_extrap**5, x_extrap**6, x_extrap**7, x_extrap**8, x_extrap**9, x_extrap**10, x_extrap**11)))

print("Predicciones para x=5:")
print(f"Grado 2: {y2[0]:.2f}")
print(f"Grado 3: {y3[0]:.2f}")
print(f"Grado 5: {y5[0]:.2f}")
print(f"Grado 7: {y7[0]:.2f}")
print(f"Grado 9: {y9[0]:.2f}")
print(f"Grado 11: {y11[0]:.2f}")
```

¿Cuál modelo crees que sea mejor?

De igual manera el modelo cuadrático, ya que captura correctamente la forma real de los datos.
